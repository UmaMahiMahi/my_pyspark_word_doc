```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number
windowSpec=Window.partitionBy("pets").orderBy("Education")
```

```python
df.withColumn("row_number", row_number().over(windowSpec)).show()
```

```python
windowSpec=Window.partitionBy("pets").orderBy("Education")
df.withColumn("row_number", row_number().over(windowSpec)).show()
```

```python
df.withColumn("new_edu",
              when(df["Education"]==1,"top").
              when(df["Education"]==2, "middle").
              when(df["Education"]==3, "primary").
              otherwise(df["Education"])).show()
```

```python
df.withColumn("exp",
              when(df["Transportation Expense"]<=50,"less").
              when((df["Transportation Expense"]>50) | (df["Transportation Expense"]<=100),"moderate").
              when(df["Transportation Expense"]>100, "high")).show()
```
